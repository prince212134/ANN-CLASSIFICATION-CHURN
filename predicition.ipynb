```python
import tensorflow as tf
from tensorflow.keras.models import load_model
import pickle 
import pandas as pd
import numpy as np
```

```python
## Load the trained model, scaler pickle,onehot
model=load_model('model.h5')

## load the encoder and scaler
with open('label_encoder_gender.pkl','rb')as file:
    label_encoder_gender=pickle.load(file)
    
    with open('one_hot_encoder.pkl','rb')as file:
     one_hot_encoder=pickle.load(file)
    
    with open('scaler.pkl','rb')as file:
     scaler=pickle.load(file)
```

```python
## Example input data

input_data ={
    	'CreditScore' : 200, 'Geography' : 'France'	, 'Gender' : 'Female' ,'Age' : 55,	'Tenure':1 ,	'Balance' : 150000,'NumOfProducts': 2,	'HasCrCard':1,	'IsActiveMember' : 0,	'EstimatedSalary' : 15000	}	
```

```python
geo_encoded=one_hot_encoder.transform([[input_data['Geography']]]).toarray()
geo_encoded_df=pd.DataFrame(geo_encoded, columns =one_hot_encoder.get_feature_names_out(['Geography']))
```

```python
geo_encoded_df
```

```python
input_df=pd.DataFrame([input_data])
input_df
```

```python
## Encode categorical variables
input_df['Gender']=label_encoder_gender.transform(input_df['Gender'])
input_df
```

```python
## concatination one hot encoded 
input_df=pd.concat([input_df.drop("Geography",axis=1),geo_encoded_df],axis=1)
input_df
```

```python
## Scaling the data

input_scaled=scaler.transform(input_df)
input_scaled
```

```python
## Predict churn

prediction=model.predict(input_scaled)
prediction
```

```python
prediction_proba=prediction[0][0]
```

```python
prediction_proba
```

```python
if prediction_proba>0.5:
    print('The customer is likely to churn.')
    
else:
    print('The customer is not likely to churn')
```
